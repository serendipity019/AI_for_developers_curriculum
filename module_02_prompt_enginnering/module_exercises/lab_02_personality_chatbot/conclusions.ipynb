## Πείραμα 2: Σύγκριση Temperature
<p style="color: green"> Έκανα το τεστ και με τα 2 μοντέλα (OpenAI και Anthropic) και παρατήρησα ότι το μοντέλο <em style="color: blue">gpt-4o-mini</em> ανταποκρίνεται ελαφρώς καλύτερα στα αναμενόμενα αποτελέσματα σύμφωνα με το temperature, σε σχέση με το <em style="color: blue">claude-sonnet-4-6</em>. <strong>Εξηγώ:</strong> Για «θερμοκρασία» 1.0, και τα δύο είχαν τα αναμενόμενα αποτελέσματα. Για «θερμοκρασία» 0.5, το <em style="color: blue">gpt-4o-mini</em> παρήγαγε και τις 2 φορές που το έτρεξα λιγότερο ισορροπημένο περιεχόμενο από το <em style="color: blue">claude-sonnet-4-6</em>, το οποίο βέβαια ήταν αρκετά ακριβές( 5 στισ 6 γραμμές ήταν ταυτόσημες), παρά απλά ισορροπημένο. Τέλος, για «θερμοκρασία» 0.0, το <em style="color: blue">gpt-4o-mini</em> είχε 5 στις 6 γραμμές το ίδιο περιεχόμενο, με την έκτη να έχει 2-3 λέξεις διαφορετικές στο τέλος της πρότασης. Το <em style="color: blue">claude-sonnet-4-6</em>, αντίστοιχα, είχε μόνο 2 στις 6 γραμμές ίδιες, ενώ στις υπόλοιπες από το μέσον της πρότασης διαφοροποιούνταν αρκετά η μία από την άλλη. <q style="color: yellow"> Συνολικά, φαίνετε ότι για ακρίβεια είναι καλύτερο το <em style="color: blue">gpt-4o-mini</em>, ή τουλάχιστον με κάνει να το εμπιστεύομαι περισσότερο για «θερμοκρασία» 0.0, σύμφωνα και με κάποια παρόμοια αποτελέσματα που είχα από άλλα τεστ στο παρελθόν μεταξύ αυτών των δύο μοντέλων. </q> </p>
