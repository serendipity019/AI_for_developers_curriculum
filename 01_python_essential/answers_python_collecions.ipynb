Exercise 2: Create a dictionary with 3 people's names and ages. Check if a name exists using in, and.get() with a default for a name that doesn't exist.


```python
my_dict = {'Alex' : 20, 'Bob' : 23, 'Cathrin' : 19}

search_name = 'Patrick'
for name in my_dict.keys():
    if name == search_name:
        print(f"The name {search_name} exists")
        print(my_dict[name])
        break
else:
    print(f"The name {search_name} didn't found")        

my_dict.get(search_name, "search name not found")    
```

Exercise 3: Create two sets of your hobbies and a friend's hobbies. Find the hobbies you share (intersection) and the ones only you have (difference).
