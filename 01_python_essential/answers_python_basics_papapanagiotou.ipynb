Exercise 2: Use input() to ask for two numbers, convert them to int, and print their sum.

```python
num_1 = int(input("Give the first integer number "))
num_2 = int(input("Give the second integer number "))
add = num_1 + num_2
print(f"The sum of {num_1} + {num_2} is {add}")
```

Exercise 3: Use chained comparison to check if your age is between 18 and 100.

```python
age = int(input("How old are you? "))
if 18 <= age <= 100:
    print("You are adult")
else:
    print("You are underaged or 'Mathousalas'")
```
