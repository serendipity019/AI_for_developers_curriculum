## Comprehensions

Exercise 1: Use a list comprehension to create a list of the first 10 cubes (1³, 2³, ..., 10³)


```python
my_nums = list(range(1, 11))
cubes = [nums ** 3 for nums in my_nums]
print(cubes)
```

Exercise 2: Given words = ["Hello", "", "World", "", "Python"], use a list comprehension to keep only non-empty strings and convert them to lowercase.


```python
given_words = ["Hello", "", "Word", "", "Python"]
new_words = [word.lower() for word in given_words if word]
print(new_words)
```

Exercise 3: Use a dict comprehension to invert a dictionary (swap keys and values). Given {"a": 1, "b": 2, "c": 3}, produce {1: "a", 2: "b", 3: "c"}.


```python
given_dict = {"a": 1, "b": 2, "c": 3}
produce_dict = {key: value for value, key in given_dict.items()}
print(produce_dict)
```

Exercise 4: Use a list comprehension with if/else to replace all negative numbers with "negative" and keep positives as-is. Input: [-5, 3, -1, 7, 0, -2].

```python
input_list = [-5, 3, -1, 7, 0, -2]
produce_list = ["negative" if num < 0 else num for num in input_list]
print(produce_list)
```
