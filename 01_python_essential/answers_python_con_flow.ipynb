Exercise 1: Write an if/elif/else that classifies a number as "positive", "negative", or "zero". Use input() to get the number.

```python
num = int(input("Give an integer number "))
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

Exercise 2: Use a for loop with range() to print the multiplication table of 7 (7×1 through 7×10).

```python
for i in range(1, 11):
    print(f"7 x {i} = {7 * i}")
```

Exercise 3: Use enumerate() over a list of your favorite foods and print each with its ranking (starting from 1).

```python
fav_food = ["Fasolada", "Pastitsio", "Mousakas", "Fakes", "Arakas"]

for i, food in enumerate(fav_food, 1):
    print(f"{i} : {food}")
```

Exercise 4: Use a while loop to keep asking for input until the user types "quit".



```python
while True:
    inp = input("Type something or 'quit' to exit")
    if inp == "quit":
        print("Goodbay")
        break
```
